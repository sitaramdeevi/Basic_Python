##Q10) Please create a new column named total_bedroom_size. If size 10 or less. It should be quoted as small. If the size is 11 or more but 1000 less, it should be medium, otherwise it should be large.

```python
housing.head() # First of all calling our existing data set
```

```python
# Creating a new column named  "total_bedroom_size"
total_bedroom_size =[6,200,900,1456,50,699,7,99,4565,138,18,1765,4,47,1997,549] * 1290 
# Multiplying it with 1290 to match the length 
housing['total_bedroom_size'] = total_bedroom_size  # ading the created  column to the data set
housing # calling the new data set  for which the created column was added.

 
```

```python
# giving the condition to the newly created column as stated in the Question
housing.loc[housing['total_bedroom_size']<=10,'bedroom_size_in_words'] = 'Small'
housing.loc[housing['total_bedroom_size']> 1000,'bedroom_size_in_words'] = 'Large'
housing # calling new data set after giving required condition

```

```python
# Checking whether the conditions were applied to the data set or not with the help of "value counts"
housing['bedroom_size_in_words'].value_counts()
```

Here we clearly see that all the required(three) conditions were applied to the data set detailed explanation is as follows


1) Number of rows whose bed_room_size '<' 10 (Small) are - 3870


2) Nnumber of rows whose bed_room_size '>' 10 and '<' 1000 (Medium) are - 11610


3) Number of rows whose bed_room_size '>' 1000 (Large) are - 5160
